```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/country_wise_latest.csv")
df.head()
```

```python
# checking if there are any null values in the data
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
# chekcing for the averge confirmed cases and recovered cases
confirmed_mean=df['Confirmed'].mean()
print("mean value of confirmed cases is:", confirmed_mean)
recovered_mean= df['Recovered'].mean()
print("mean value of recovered cases is:", recovered_mean)

# check for the confirmed and recovered ratio
df['RecoveryRatio'] = df["Confirmed"] / df['Recovered']
print(df['RecoveryRatio'])
```

```python
df.sort_values('Confirmed',ascending=False)
```

```python
x = df['Confirmed']
y = df['Recovered']
plt.scatter(x,y)
plt.xlabel("Confiremd Cases")
plt.ylabel("recovered Cases")
plt.title("Visualization of Confirmed and Recovered cases")
plt.show()
```

```python
x = df['Confirmed']
y = df['Country/Region']
plt.plot(x,y, color ='r')
plt.plot(df['Recovered'],y,color='g')
plt.legend(['confirmed','Recovered'])
plt.xlabel("Confiremd Cases")
plt.ylabel("Country/Region")
plt.title("Visualization of Confirmed and Recovered cases")
plt.show()
```

```python
plt.plot(df['Deaths'],df['WHO Region'],color='r')
plt.plot(df['Recovered'],df['WHO Region'],color='g')
plt.title('The death and recovery rate')
plt.xlabel('Count')
plt.ylabel('WHO Region')
plt.legend(['Deaths','Recovered'])
plt.show()
```

```python
x1=df['New deaths']
y1=df['New recovered']
bars=plt.bar(x1,y1,width=50,color='#abcdef',edgecolor='c')
plt.title('The New deaths and New recovered data')
plt.xlabel('New deaths')
plt.ylabel('New recovered')
plt.legend(['New deaths','New recovered'])
plt.show()
```
